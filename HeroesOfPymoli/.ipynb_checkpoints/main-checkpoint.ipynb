```python
#importing dependencies

import pandas as pd

#reading in csv file

purchase_data = pd.read_csv('../Resources/purchase_data.csv')

purchase_data.head()

purchase_data.set_index('Purchase ID')
```

```python
#Determining number of players and displaying it
player_count_df = pd.DataFrame({'Total Players': [len((purchase_data['SN'].unique()))]})
```

```python
player_count_df
```

```python
items = len(purchase_data['Item Name'].unique())
avg_price = purchase_data['Price'].mean()
num_of_purchases = len(purchase_data['Item Name'])
total_revenue = purchase_data['Price'].sum()

purchase_analysis_df = pd.DataFrame({
    'Number of Unique Items':[items],
    'Average Price':[avg_price],
    'Number of Purchases': [num_of_purchases],
    'Total Revenue':[total_revenue]
})

purchase_analysis_df
```

```python
unique = []
male = 0
female = 0
other = 0


for column, row in purchase_data.iterrows():
    if row['SN'] not in unique:
        if row['Gender'] == 'Male':
            male +=1
        elif row['Gender'] == 'Female':
            female +=1
        else:
            other +=1
        unique.append(row['SN'])

total = male + female + other
male_per = male / total * 100
female_per = female / total * 100
other_per = other / total * 100

gender_df = pd.DataFrame({
    None : ['Male','Female', 'Other / Non-Disclosed'],
    'Total Count' : [male, female, other],
    'Percentage of Players' : [male_per, female_per, other_per]
})

gender_df.set_index(None)
```

```python
gender_groupby = purchase_data.groupby(['Gender'])

purchase_count_gender = gender_groupby['SN'].count()

avg_price_gender = gender_groupby['Price'].mean()

total_purchase_gender = gender_groupby['Price'].sum()

total_purchase_gender

gender_df.rename(columns={None:'Gender'}, inplace = True)


purchasing_analysis_gender_df = pd.DataFrame({
    'Purchase Count' : purchase_count_gender[0:3],
    'Average Purchase Price' : avg_price_gender[0:3],
    'Total Purchase Value' : total_purchase_gender[0:3],
})

purchasing_analysis_gender_df

merge_gender_df = pd.merge(purchasing_analysis_gender_df, gender_df, on = 'Gender')


merge_gender_df['Avg Total Purchase Per Person'] = merge_gender_df['Total Purchase Value'] / merge_gender_df['Total Count']

merge_gender_df

merge_gender_df.drop(columns = ['Total Count', 'Percentage of Players'], inplace = True)

merge_gender_df.set_index(['Gender'])
```

```python
age_bins = [0,9,14,19,24,29,34,39,150]
age_group_names = ['<10', '10-14','15-19','20-24','25-29','30-34','35-39','40+']

unique_players_df = purchase_data.drop_duplicates(subset = ['SN'])

unique_players_df['Age Group'] = pd.cut(unique_players_df['Age'], age_bins, labels = age_group_names, include_lowest=True)

unique_players_df.head()

total_players = len(unique_players_df['SN'])

age_groupby = unique_players_df.groupby(['Age Group'])

total_age_count = age_groupby.count()



age_count_df = pd.DataFrame({
    #None: age_group_names,
    'Total Count' : total_age_count['Age']
})

age_count_df['Percentage of Players'] = round(age_count_df['Total Count'] / total_players * 100, 2)


#age_count_df.set_index(None)

age_count_df
```

```python
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], age_bins, labels = age_group_names, include_lowest=True)

full_age_groupby = purchase_data.groupby(['Age Group'])

age_purchase_count = full_age_groupby['SN'].count()

age_avg_purchase_price = round(full_age_groupby['Price'].mean(),2)

age_total_purchase_value = round(full_age_groupby['Price'].sum(),2)

age_total_purchase_value

purchase_analysis_age_df = pd.DataFrame({
    'Purchase Count' : age_purchase_count[:],
    'Average Purchase Price' : age_avg_purchase_price[:],
    'Total Purchase Value' : age_total_purchase_value[:]
})



merge_age_purchase_df = pd.merge(age_count_df, purchase_analysis_age_df, on = 'Age Group')


merge_age_purchase_df['Avg Total Purchase Per Person'] = round(merge_age_purchase_df['Total Purchase Value'] / merge_age_purchase_df['Total Count'], 2)

merge_age_purchase_df

merge_age_purchase_df.drop(columns = ['Total Count', 'Percentage of Players'], inplace = True)

merge_age_purchase_df

```

```python
top_groupby = purchase_data.groupby(['SN'])

purchase_count_top = top_groupby['Age'].count()

#avg_purchase_price_top = 

total_purchase_value_top = top_groupby['Price'].sum()

top_spenders_df = pd.DataFrame({
    'Purchase Count' : purchase_count_top[:],
    'Total Purchase Value' : total_purchase_value_top[:]
})


top_spenders_df['Average Purchase Price'] = round(top_spenders_df['Total Purchase Value'] / top_spenders_df['Purchase Count'], 2)


top_spenders_df = top_spenders_df[['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']]
top_spenders_df.sort_values(by='Total Purchase Value', ascending=False, inplace =True)

top_spenders_df.head()
```

```python
pop_groupby = purchase_data.groupby('Item Name')

pop_purchase_count = pop_groupby['SN'].count()

pop_total_purchase_value = pop_groupby['Price'].sum()

pop_items_df = pd.DataFrame({
    'Purchase Count' : pop_purchase_count[:],
    'Total Purchase Value' : pop_total_purchase_value[:]
})



pop_ID_price_df = purchase_data[['Item ID', 'Item Name']]
pop_ID_price_df.drop_duplicates(inplace =True)
pop_ID_price_df.sort_index(inplace = True)

pop_ID_price_df.head()



merge_pop_df = pd.merge(pop_items_df, pop_ID_price_df, on='Item Name')
merge_pop_df.head(5)
merge_pop_df['Item Price'] = round(merge_pop_df['Total Purchase Value'] / merge_pop_df['Purchase Count'], 2)

merge_pop_df = merge_pop_df[['Item ID', 'Item Name', 'Purchase Count', 'Item Price', 'Total Purchase Value']]


merge_pop_df.sort_values(by='Purchase Count',ascending=False, inplace =True)
merge_pop_df.set_index(['Item ID', 'Item Name'], inplace= True)
#merge_pop_df.reset_index(drop = True, inplace = True)


merge_pop_df.head(10)
```

```python
most_prof_df = merge_pop_df

most_prof_df.sort_values(by='Total Purchase Value', ascending=False, inplace =True)

most_prof_df.head()
```
